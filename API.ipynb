```python
import requests,json
def get_coordinates():
    api_key='9e0ab0f27fbede4d479191d334dde557'
    city=input("Enter City Name : ")
    request_url = f" http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response = requests.get(request_url)
    coordinates = response.json()
    lattitude = coordinates[0]['lat']
    longitude = coordinates[0]['lon']
    request_url1= f" http://api.openweathermap.org/data/2.5/forecast?lat={lattitude}&lon={longitude}&appid={api_key}"
    response1 = requests.get(request_url1)
    d5_data = response1.json()
    print(json.dumps(d5_data,indent=5))
    return d5_data,city
wd,city=get_coordinates()
```

```python
date=[]
time=[]
temp=[]
humidity=[]
pressure=[]
wed=[]
for i in range(0,40):
    dt=wd['list'][i]['dt_txt']
    date.append(dt[0:10])
    time.append(dt[11:])
    temp.append(wd['list'][i]['main']['temp'])
    pressure.append(wd['list'][i]['main']['pressure'])
    humidity.append(wd['list'][i]['main']['humidity'])
    wed.append(wd['list'][i]['weather'][0]['description'])
    print(dt)
```
